# 01 — Data Download

Download 5 years of daily OHLCV data for 20 liquid NSE large-cap stocks. This notebook
only downloads and validates raw data — no feature engineering or EDA here.

```python
import yfinance as yf
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
TICKERS = [
    "RELIANCE.NS", "TCS.NS", "INFY.NS", "HDFCBANK.NS", "ICICIBANK.NS",
    "HINDUNILVR.NS", "ITC.NS", "SBIN.NS", "BHARTIARTL.NS", "KOTAKBANK.NS",
    "LT.NS", "AXISBANK.NS", "ASIANPAINT.NS", "MARUTI.NS", "SUNPHARMA.NS",
    "TITAN.NS", "ULTRACEMCO.NS", "WIPRO.NS", "NESTLEIND.NS", "TATAMOTORS.NS",
]
START = "2019-01-01"
END   = "2024-01-01"

# Yahoo Finance retired TATAMOTORS.NS following Tata Motors' 2025 demerger into
# commercial-vehicle and passenger-vehicle entities. The passenger-vehicle symbol
# TMPV.NS carries forward the same historical price series for our date range, so
# we fall back to it and still save the output under the original ticker name.
YF_SYMBOL_OVERRIDES = {"TATAMOTORS.NS": "TMPV.NS"}

raw = {}
for ticker in TICKERS:
    yf_symbol = YF_SYMBOL_OVERRIDES.get(ticker, ticker)
    df = yf.download(yf_symbol, start=START, end=END, auto_adjust=True, progress=False)
    if isinstance(df.columns, pd.MultiIndex):
        df.columns = [col[0] for col in df.columns]
    df.dropna(inplace=True)
    raw[ticker] = df
    print(f"{ticker}: {df.shape[0]} rows | {df.index[0].date()} to {df.index[-1].date()}")
```

## Sanity checks

```python
for ticker, df in raw.items():
    print(f"{ticker:15s} shape={df.shape}  range={df.index[0].date()} -> {df.index[-1].date()}  missing={df.isnull().sum().sum()}")
```

## Save raw data

```python
for ticker, df in raw.items():
    fname = ticker.replace(".", "_")
    df.to_csv(f"../data/{fname}_raw.csv")
    print(f"Saved ../data/{fname}_raw.csv")
```
